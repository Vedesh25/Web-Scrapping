```python
import requests 
```

```python
import bs4
```

```python
url="https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city"
```

```python
data=requests.get(url)
data
```

```python
type(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(data.text,'html')
```

```python
soup.find('div',attrs={"class":"title-line"}).text
```

```python
soup.find('span',attrs={'itemprop':"addressLocality"}).text
```

```python
soup.find('div',attrs={'data-type':"price-link"}).text
```

```python
soup.find('td',attrs={'class':"size"}).text
```

```python
soup.find('td',attrs={'class':"val"}).text
```

```python
soup.find('td',attrs={"class":"lbl rate"}).text
```

```python
a=soup.find_all('div',attrs={"class":"title-line"})
prop_name=[i.text for i in a]
print(len(prop_name))
print(prop_name)
```

```python
b=soup.find_all('span',attrs={'itemprop':"addressLocality"})
location=[i.text for i in b]
print(len(location))
print(location)
```

```python
c=soup.find_all('div',attrs={'data-type':"price-link"})
price=[i.text for i in c]
print(len(price))
print(price)
```

```python
d=soup.find_all('td',attrs={'class':"size"})
year=[i.text for i in d]
print(len(year))
print(year)
```

```python
e=soup.find_all('td',attrs={'class':"val"})
uc=[i.text for i in e]
print(len(uc))
print(uc)
```

```python
f=soup.find_all('td',attrs={"class":"lbl rate"})
sqf=[i.text for i in f]
print(len(sqf))
print(sqf)
```

```python
import pandas as pd
```

```python
data_1=pd.DataFrame({'prop_name':prop_name,'location':location,'price':price,'year':year,'uc':uc,'sqf':sqf})
data_1
```

```python
data_1.to_csv('Hyd_aparts_web_scrapping.csv')
```
